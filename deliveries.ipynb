### replacing the team name with the abbreviations

```python


Teams={
    'Royal Challengers Bangalore':'RCB', 
    'Sunrisers Hyderabad':'SRH',
       'Rising Pune Supergiant':'RPS',
    'Mumbai Indians':'MI',
       'Kolkata Knight Riders':'KKR', 
    'Gujarat Lions':'GL',
    'Kings XI Punjab':'KXIP',
       'Delhi Daredevils':'DD',
    'Chennai Super Kings':'CSK',
    'Rajasthan Royals':'RR',
       'Deccan Chargers':'DC',
    'Kochi Tuskers Kerala':'KTK',
    'Pune Warriors':'PW',
       'Rising Pune Supergiants':'RPS'
}
```

```python
data['batting_team'] = data['batting_team'].map(Teams)
```

```python
data['bowling_team'] = data['bowling_team'].map(Teams)
```

# SCORE DISTRIBUTION FOR THE TEAMS BY THE INNINGS

```python
runs=data.groupby(['match_id','inning','batting_team'])[['total_runs']].sum().reset_index()
runs.drop('match_id',axis=1,inplace=True)
```

```python
runs.head(10)
```

```python
inning1 = runs[runs['inning']==1]
inning2 = runs[runs['inning']==2]
```

```python
runs.head()
```

```python
sns.boxplot(x='batting_team',y='total_runs',data=inning1)
```

## ACCORDING TO THIS PLOT, THE BATTING BY CSK AND RCB SEEMS BEST

```python
sns.boxplot(x='batting_team',y='total_runs',data=inning2)
```

```python
# How many times team scored more than 200.
high_scores = data.groupby(['match_id','inning','batting_team','bowling_team'])['total_runs'].sum().reset_index()
```

```python
high_scores
```

```python
score_200 = high_scores[high_scores['total_runs']>=200]
score_200
```

```python
plt.figure(figsize=(10,6))
sns.countplot(score_200['batting_team'])
```
